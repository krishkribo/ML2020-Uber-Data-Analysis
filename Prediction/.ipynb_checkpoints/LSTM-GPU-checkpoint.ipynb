```python
import numpy as np
import pandas as pd
import nbimporter
import geopandas as gpd
import commons.utility.pickle_util as pickle
import Preprocessing.SpatialMapping as sm
import Preprocessing.Discretization as dis
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import time
from pandas.plotting import register_matplotlib_converters
#register_matplotlib_converters()
import matplotlib.pyplot as plt

coord_system = {'init': 'epsg:4326'}
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
from keras.preprocessing.sequence import TimeseriesGenerator
import keras
```

```python
data = sm.run(['apr', 'may'])
```

```python
df = data.copy()
```

```python
df2 = df[['datetime', 'zone']]
df2g = df2.groupby(['datetime', 'zone']).size().reset_index(name='counts')
df2g = dis.encode(df2g, ['zone'])
df2g['datetime'] = pd.to_datetime(df2g['datetime'], format='%m-%d-%Y:%H').apply(lambda x: time.mktime(x.timetuple()))
```

```python
test_size = int(len(df2g) * 0.1) # the test data will be 10% (0.1) of the entire data
train = df2g.iloc[:-test_size,:].copy() 
# the copy() here is important, it will prevent us from getting: SettingWithCopyWarning: A value is trying to be set on a copy of a slice from a DataFrame. Try using .loc[row_index,col_indexer] = value instead
test = df2g.iloc[-test_size:,:].copy()
print(train.shape, test.shape)
```

```python
plt.figure(figsize=(50,4))
plt.plot(train.index,train['counts'],label='Train');
plt.plot(test.index,test['counts'],label='test')
plt.legend();
```

```python
y_col = 'counts'
```

```python
#separate X and y only for the train data (for now)
X_train = train.drop(y_col,axis=1).copy()
y_train = train[[y_col]].copy() # the double brakets here are to keep the y in a dataframe format, otherwise it will be pandas Series
print(X_train.shape, y_train.shape)
```

```python
Xscaler = MinMaxScaler(feature_range=(0, 1)) # scale so that all the X data will range from 0 to 1
Xscaler.fit(X_train)
scaled_X_train = Xscaler.transform(X_train)
print(X_train.shape)
Yscaler = MinMaxScaler(feature_range=(0, 1))
Yscaler.fit(y_train)
scaled_y_train = Yscaler.transform(y_train)
print(scaled_y_train.shape)
scaled_y_train = scaled_y_train.reshape(-1) # remove the second dimension from y so the shape changes from (n,1) to (n,)
print(scaled_y_train.shape)

scaled_y_train = np.insert(scaled_y_train, 0, 0)
scaled_y_train = np.delete(scaled_y_train, -1)
```

```python
tmpXtrainDf = pd.DataFrame(scaled_X_train, columns=['datetime', 'zone'])
tmpXtrainDf['datetime'] = X_train['datetime']
scaled_X_train2 = tmpXtrainDf.to_numpy()
```

```python

n_input = 25 #how many samples/rows/timesteps to look in the past in order to forecast the next sample
n_features= X_train.shape[1] # how many predictors/Xs/features we have to predict y
b_size = 32 # Number of timeseries samples in each batch
generator = TimeseriesGenerator(scaled_X_train, scaled_y_train, length=n_input, batch_size=b_size)

print(generator[0][0].shape)
```

```python
x_train = X_train.to_numpy().reshape(1, X_train.shape[0], X_train.shape[1]) #ok shape (1,steps,5) - 1 sequence, many steps, 5 features
y_train = y_train.to_numpy().reshape(1, y_train.shape[0], y_train.shape[1])
# x_test = X_test.to_numpy().reshape(1, X_test.shape[0], X_test.shape[1])
# y_test = Y_test.to_numpy().reshape(1, Y_test.shape[0], Y_test.shape[1])
```

```python
model = Sequential()
model.add(LSTM((20), input_shape=(n_input, n_features), return_sequences=True))
model.add(LSTM(20, return_sequences=False))
model.add(Dense(1, activation='relu'))
model.compile(loss='mse', optimizer='adam')
```

```python
# from keras.datasets import mnist
```

```python
# (X_train, y_train),(X_test, y_test) = mnist.load_data() 
```

```python
# with tf.device("cpu:0"):
with tf.device("cpu:0"):
    model.fit_generator(generator, epochs=50)
```

```python
model.save('../data/picklerCheckpoints/LSTM/relu50')
```

```python
loss_per_epoch = model.history.history['loss']
plt.plot(range(len(loss_per_epoch)),loss_per_epoch);
```

```python
X_test = test.drop(y_col,axis=1).copy()
scaled_X_test = Xscaler.transform(X_test)
test_generator = TimeseriesGenerator(scaled_X_test, np.zeros(len(X_test)), length=n_input, batch_size=b_size)
print(test_generator[0][0].shape)
```

```python
with tf.device("cpu:0"):
    y_pred_scaled = model.predict(test_generator)
y_pred = Yscaler.inverse_transform(y_pred_scaled)
results = pd.DataFrame({'y_true':test[y_col].values[n_input:],'y_pred':y_pred.ravel()})
print(results)
```
